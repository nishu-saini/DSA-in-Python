```python
# Linked List implementation..

# class to store each node information..
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

# linked list class..
class linkedList:
    def __init__(self):
        self.head = None
        
    # method to insert node at the beginning..
    def insert_at_beginning(self, data):
        node = Node(data, self.head)
        self.head = node
        
    # method to print linked list..
    def printll(self):
        # check if head is None..
        if self.head is None:
            print('Linked list is empty')
            return
        
        # store node..
        itr = self.head
        llstr = ''
        
        while itr:   # until itr in not None.. 
            llstr += str(itr.data) + '-->'
            itr = itr.next
            
        print(llstr)
        
    # method to insert node at the end..
    def insert_at_end(self, data):
        # check if head is None
        if self.head is None:
            # means ll is empty..
            # first element will go at end
            self.head = Node(data, None)
            return
            
        itr = self.head
        
        while itr.next:
            # until itr gets None
            itr = itr.next
        
        # itr is None, Now add node at end..
        itr.next = Node(data, None)
        
    # method to create fresh ll when a list of data is given..
    def insert_values(self, data_list):
        # make head None or empty..
        self.head = None
        
        for data in data_list:
            self.insert_at_end(data)
            
    # method to get length of ll..
    def get_length(self):
        length = 0
        itr = self.head
        
        while itr:
            length += 1
            itr = itr.next  
            
        return length
    
    # method to remove node when particular index is given..
    def remove_at(self, index):
        # handle exception for given index
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid Index")
        
        # if given index is 0..
        if index == 0:
            self.head = self.head.next
            return
        
        indx = 0
        itr = self.head
        
        while itr:
            if indx == index - 1:   # prev node of given index
                # assign next node to it's next node
                itr.next = itr.next.next
                break
                
            itr = itr.next
            indx += 1
            
    # method to insert at given index..
    def insert_at(self, index, data):
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid Index")
            
        if index == 0:
            self.insert_at_beginning(data)
            return
        
        indx = 0
        itr = self.head
        
        while itr:
            if indx == index - 1:
                node = Node(data, itr.next)
                itr.next = node
                break
                
            itr = itr.next
            indx += 1
    
    
```
